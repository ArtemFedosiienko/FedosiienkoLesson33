# Розв'язок #

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 1. Створення датафрейму з xls-документу: ####

```python
df_cars = pd.read_excel('cars.xlsx')
df_cars
```

#### 2. Візуалізація залежності: ####

##### Залежність споживання палива від вартості #####

```python
sns.set_style('darkgrid')
plt.plot(sorted(df_cars['Cost']), sorted(df_cars['Oil']), color = 'blue', linewidth = 3)
plt.xlabel('Вартість авто, $')
plt.ylabel('Споживання палива, л/100км')
```

##### Залежність споживання палива від швидкості #####

```python
plt.scatter(sorted(df_cars['Max_Speed']), sorted(df_cars['Oil']), color = 'blue', marker = 'D')
plt.xlabel('Швидкість, км/год')
plt.ylabel('Споживання палива, л/100км')
```

##### Залежність вантажопідйомності від швидкості #####

```python
plt.scatter(sorted(df_cars['Max_Speed']), sorted(df_cars['Load_capacity']), color = 'blue', marker = '^')
plt.xlabel('Швидкість, км/год')
plt.ylabel('Вантажопідйомність, кг')
```

#### 3. Побудова гістограми: ####

```python
plt.hist(df_cars['Max_Speed'], 5, color = 'blue', width = 13.5)
plt.xlabel('Швидкість, км/год')
plt.ylabel('Марки авто, шт.')
```

#### 4. Побудова коробкового графіку: ####

```python
sns.boxplot(x='Max_Speed', y='Car', data=df_cars)
```

#### 4. Побудова парного графіку: ####

```python
sns.pairplot(df_cars)
```
